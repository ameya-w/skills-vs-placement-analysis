```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'D:\\Coding\\Pandas\\skill-vs-placement.csv', index_col='student_id')
```

```python
print(df)
```

```python
placed_df = df[df['placement_status'] == 'Placed']
placed_median = placed_df.groupby('internships')['package_lpa'].median()
print(placed_median)
plt.figure(figsize=(12,6))
plt.bar(placed_median.index , placed_median.values)
plt.xlabel("Number of Internships")
plt.ylabel("Median Package in LPA")
plt.show()
```

```python
branch_package = df.loc[df['placement_status'] == 'Placed', ['branch', 'package_lpa']]
branch_median = branch_package.groupby('branch')['package_lpa'].median()
print(branch_package)
print(branch_median)
plt.figure(figsize=(12,6))
plt.bar(branch_median.index, branch_median.values)
plt.xlabel("Branch")
plt.ylabel("Median Package in LPA")
plt.show()
```

```python
branch_intern = df.loc[df['placement_status'] == 'Placed', ['branch', 'internships', 'job_role', 'package_lpa']]
branch_median = branch_intern.groupby('branch')['package_lpa'].median()
role_median = branch_intern.groupby('job_role')['package_lpa'].median()
plt.figure(figsize=(12,6))
plt.bar(role_median.index, role_median.values)
plt.xlabel("Job Role")
plt.ylabel("Median Package in LPA")
plt.show()
```
